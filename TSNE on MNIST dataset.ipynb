```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

D=pd.read_csv(r'C:\Users\dell\Downloads\train.csv')
print(D.head(5))
l=D['label']
d=D.drop("label",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
standerdized_data=StandardScaler().fit_transform(d)
print(standerdized_data.shape)
```

```python
from sklearn.manifold import TSNE

model=TSNE(n_components=2,random_state=0,perplexity=50,n_iter=5000)

tsne_data=model.fit_transform(standerdized_data)
tsne_data=np.vstack((tsne_data.T,l)).T
tsne_df=pd.DataFrame(data=tsne_data,columns=("Dim_1","Dim_2","label"))
sns.FacetGrid(tsne_df,hue="label",height=6).map(plt.scatter,'Dim_1','Dim_2','label').add_legend()
plt.show()
```
